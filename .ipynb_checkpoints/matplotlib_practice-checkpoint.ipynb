```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
#visual sample
savings = [0,100,150,200,300,500,300,350,500,600,750,1000]
months = ['Jan','Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
plt.plot(months,savings)
plt.xlabel('Month')
plt.ylabel('Savings($)')
plt.title('Savings Account Results')
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot (vertical bar chart)
x = x_axis
y = y_axis
plt.bar(x_axis, y_axis, color='green', label='Boston')
# Create X and Y axis labels
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
# Create the plot (horizontal bar chart)
plt.barh(x_axis, y_axis, color='m', label='Boston')
# Create X and Y axis labels
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Create a title
plt.title("PyBer Fare by Month")
# invert months
plt.gca().invert_yaxis()
# Add the legend
plt.legend()
```

```python
# Create the plot with ax.plot()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="black", label='New York')
```

```python
# Create the plot with ax.plot (horizontal bar)
fif, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="c", label='Chicago')
ax.set_title("PyBer Fare by Month")
plt.xlabel("Fares($)")
plt.ylabel("Dates")
ax.invert_yaxis()
```

```python
# Create a scatter plot plt.plot method
plt.plot(x_axis, y_axis,'o')
```

```python
# Create a scatter plot(scatter method)
plt.scatter(y_axis, x_axis, color="red", label="Chicago")
plt.xlim(1,50)
plt.xlabel("Fares($)")
plt.ylabel("Date")
plt.title("PyBer by Month")
plt.gca().invert_yaxis()
```

```python
# Create scatter plot with sized bubbles

y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#same as above but on 1 line without for loop
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], color="skyblue", label="Boston", edgecolor ="k", linewidths =2, norm =.1 )
plt.xlabel("Fares($)")
plt.ylabel("Date")
plt.title("PyBer by Month")
plt.legend()
plt.xlim(1,50)
ax.invert_yaxis()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8,8))

plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
explode2 = [0,0,0.2,0,0,0,0.2,0,0,0,0,0]
colors=["slateblue", "magenta", "lightblue", "green", "pink", "greenyellow", "yellow", "orange", "gold", "purple", "tomato", "mistyrose"]

fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis, 
       startangle=95, 
       shadow= True, 
       counterclock=False, 
       colors = colors,
       explode = explode2, 
       autopct='%.1f%%')
fig = plt.gcf()
fig.set_size_inches(8,8)
plt.show()
```
