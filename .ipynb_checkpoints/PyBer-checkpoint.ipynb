```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and setup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Files to load
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"
```

```python
city_data_df = pd.read_csv(city_data_to_load)
city_data_df.head(10)
```

```python
ride_data_df = pd.read_csv(ride_data_to_load)
ride_data_df.head(10)
```

```python
# Get the columns and rows that are not null
city_data_df.count()
```

```python
# Check that there are no null values in city data
city_data_df.isnull().sum()
```

```python
# Get the data types of each column
city_data_df.dtypes
```

```python
# Get the unique values of the type of city
city_data_df["type"].unique()
```

```python
# Get the number of data points from the Urban Cities.
sum(city_data_df["type"]=="Urban")
```

```python
# Get the number of data points from the Suburban Cities.
sum(city_data_df["type"]=="Suburban")
```

```python
# Get the number of data points from the Rural Cities.
sum(city_data_df["type"]=="Rural")
```

```python
# Get the columns and rows that are not null in ride data
ride_data_df.count()
```

```python
# Check that there are no null values in ride data
ride_data_df.isnull().sum()
```

```python
# Get the data types of each column in ride data
ride_data_df.dtypes
```

```python
# Combine into a single dataset
pyber_data_df =pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])
# Show the combined DataFrame
pyber_data_df.head(10)
```

```python
# Create the Urban city DataFrame
urban_cities_df =pyber_data_df[pyber_data_df["type"]=="Urban"]
urban_cities_df.head()
```

```python
# Create the Suburban city DataFrame
suburban_cities_df =pyber_data_df[pyber_data_df["type"]=="Suburban"]
suburban_cities_df.head()
```

```python
# Create the Rural city DataFrame
rural_cities_df =pyber_data_df[pyber_data_df["type"]=="Rural"]
rural_cities_df.head()
```

```python
# Get the number of rides for urban cities.
urban_ride_count = urban_cities_df. groupby(["city"]).count()["ride_id"]
urban_ride_count.head()
```

```python
# Get the number of rides for suburban cities.
suburban_ride_count = suburban_cities_df.groupby(["city"]).count()["ride_id"] 
suburban_ride_count.head()
```

```python
# Get the number of rides for rural cities. 
rural_ride_count = rural_cities_df.groupby(["city"]).count()["ride_id"]
rural_ride_count.head()
```

```python
# Get the average fare for each urban city. 
urban_avg_fare = urban_cities_df.groupby(["city"]).mean()["fare"]
urban_avg_fare.head()
```

```python
# Get the average fare for each suburban city. 
suburban_avg_fare = suburban_cities_df.groupby(["city"]).mean()["fare"]
suburban_avg_fare.head()
```

```python
# Get the average fare for each rural city. 
rural_avg_fare = rural_cities_df.groupby(["city"]).mean()["fare"]
rural_avg_fare.head()
```

```python
# Get the average number of drivers in each urban city
urban_driver_count = urban_cities_df.groupby(["city"]).mean()["driver_count"]
urban_driver_count.head()
```

```python
# Get the average number of drivers in each suburban city
suburban_driver_count = suburban_cities_df.groupby(["city"]).mean()["driver_count"]
suburban_driver_count.head()
```

```python
# Get the average number of drivers in each rural city
rural_driver_count = rural_cities_df.groupby(["city"]).mean()["driver_count"]
rural_driver_count.head()
```

```python
# Build Urban cities scatter plots.
plt.scatter(urban_ride_count, urban_avg_fare)
```

```python
# Build the scatter plots for urban cities. 
plt.scatter(urban_ride_count,
           urban_avg_fare,
           s= urban_driver_count)
```

```python
# Build the scatter plots for urban cities. 
plt.scatter(urban_ride_count,
           urban_avg_fare,
           s= 10*urban_driver_count,
           color="coral",
           edgecolor="k",
           linewidth=1,
           alpha=.8,
           label="Urban")
plt.title("PyBer Ride-Sharing Data(2019)")
plt.xlabel("Total Number of Riders (Per City)")
plt.ylabel("Average Fare($)")
plt.grid()
plt.legend()
```

```python
# Build the scatter plots for suburban cities. 
plt.scatter(suburban_ride_count,
           suburban_avg_fare,
           s= 10*suburban_driver_count,
           color="skyblue",
           edgecolor="k",
           linewidth=1,
           alpha=.8,
           label="Suburban")
plt.title("PyBer Ride-Sharing Data(2019)")
plt.xlabel("Total Number of Riders (Per City)")
plt.ylabel("Average Fare($)")
plt.grid()
plt.legend()
```

```python
# Build the scatter plots for rural cities. 
plt.scatter(rural_ride_count,
           rural_avg_fare,
           s= 10*rural_driver_count,
           color="gold",
           edgecolor="k",
           linewidth=1,
           alpha=.8,
           label="Rural")
plt.title("PyBer Ride-Sharing Data(2019)")
plt.xlabel("Total Number of Riders (Per City)")
plt.ylabel("Average Fare($)")
plt.grid()
plt.legend()
```

```python
# Add scatter plots for each type of city. 
#Urban
plt.scatter(urban_ride_count,
           urban_avg_fare,
           s= 10*urban_driver_count,
           color="coral",
           edgecolor="k",
           linewidth=1,
           alpha=.8,
           label="Urban")

#Suburban
plt.scatter(suburban_ride_count,
           suburban_avg_fare,
           s= 10*suburban_driver_count,
           color="skyblue",
           edgecolor="k",
           linewidth=1,
           alpha=.8,
           label="Suburban")

#Rural
plt.scatter(rural_ride_count,
           rural_avg_fare,
           s= 10*rural_driver_count,
           color="gold",
           edgecolor="k",
           linewidth=1,
           alpha=.8,
           label="Rural")
# Show plot
plt.show()
```

```python
# Add scatter plots for each type of city. 
#Urban
plt.scatter(urban_ride_count,
           urban_avg_fare,
           s= 10*urban_driver_count,
           color="coral",
           edgecolor="k",
           linewidth=1,
           alpha=.8,
           label="Urban")

#Suburban
plt.scatter(suburban_ride_count,
           suburban_avg_fare,
           s= 10*suburban_driver_count,
           color="skyblue",
           edgecolor="k",
           linewidth=1,
           alpha=.8,
           label="Suburban")

#Rural
plt.scatter(rural_ride_count,
           rural_avg_fare,
           s= 10*rural_driver_count,
           color="gold",
           edgecolor="k",
           linewidth=1,
           alpha=.8,
           label="Rural")

# Other graph properties added
plt.title("PyBer Ride-Sharing Data(2019)", fontsize=20)
plt.xlabel("Total Number of Riders (Per City)",fontsize=12)
plt.ylabel("Average Fare($)",fontsize=12)
plt.grid(True)
plt.legend()
# Show plot
plt.show()
```

```python
# Add scatter plots for each type of city. 
#Urban
plt.scatter(urban_ride_count,
           urban_avg_fare,
           s= 10*urban_driver_count,
           color="coral",
           edgecolor="k",
           linewidth=1,
           alpha=.8,
           label="Urban")

#Suburban
plt.scatter(suburban_ride_count,
           suburban_avg_fare,
           s= 10*suburban_driver_count,
           color="skyblue",
           edgecolor="k",
           linewidth=1,
           alpha=.8,
           label="Suburban")

#Rural
plt.scatter(rural_ride_count,
           rural_avg_fare,
           s= 10*rural_driver_count,
           color="gold",
           edgecolor="k",
           linewidth=1,
           alpha=.8,
           label="Rural")

# Other graph properties added
plt.title("PyBer Ride-Sharing Data(2019)", fontsize=20)
plt.xlabel("Total Number of Riders (Per City)",fontsize=12)
plt.ylabel("Average Fare($)",fontsize=12)
plt.grid(True)
# Create the legend
lgnd = plt.legend(fontsize="12", mode="Expanded", scatterpoints=1, loc="best", title= "City Types")
lgnd.legendHandles[0].sizes =[75]
lgnd.legendHandles[1].sizes =[75]
lgnd.legendHandles[2].sizes =[75]
lgnd.get_title().set_fontsize(12)
#Incorporate a text label about circle size. 
plt.text(42, 35, "Note: Circle Size correlates with driver count per city.", fontsize="12")
# Show plot
plt.show()
```
